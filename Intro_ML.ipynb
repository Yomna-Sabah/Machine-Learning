### Question 1
Pandas dataframes are much easier and more interpretable when dealing with tables. **Convert the cancer data into dataframe.**

Note: Scikit-learn works with lists, numpy arrays, scipy-sparse matrices, and pandas DataFrames, so converting the dataset to a DataFrame is not necessary for training any model.

*This function should return a (569, 31) DataFrame with *

*columns = *

    ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
    'mean smoothness', 'mean compactness', 'mean concavity',
    'mean concave points', 'mean symmetry', 'mean fractal dimension',
    'radius error', 'texture error', 'perimeter error', 'area error',
    'smoothness error', 'compactness error', 'concavity error',

    'worst radius', 'worst texture', 'worst perimeter', 'worst area',

    'worst concave points', 'worst symmetry', 'worst fractal dimension',
    'target']
    
*and index = *

RangeIndex(start=0, stop=569, step=1)

```python
def answer_one():
    
    # Your code here
    # create cancer DataFrame
    cancerdf = pd.DataFrame(cancer.data, columns = cancer.feature_names, dtype = float)
    # create target DataFrame
    targetdf = pd.DataFrame(cancer.target, columns = ['target'])
    # Join column of target DataFrame to cancer DataFrame
    cancerdf = cancerdf.join(targetdf)
    
    return cancerdf# Return your answer


answer_one()
```

### Question 2
What is the class distribution? (i.e. how many instances of `malignant` and how many `benign`?)

```python
def answer_two():
    cancerdf = answer_one()
    
    # Your code here
    
    # In target column in cancer DataFrame replace 0 with malignant and 1 with benign
    cancerdf['target'] = cancerdf['target'].replace([0, 1], ['malignant', 'benign'])
    # count no. of malignant and benign in target column
    counts = cancerdf['target'].value_counts()
    # return target column in cancer DataFrame as it were
    cancerdf['target'] = cancerdf['target'].replace(['malignant', 'benign'], [0, 1])
    
    return counts# Return your answer


answer_two()
```

### Question 3
Split the DataFrame into `X` (the data) and `y` (the labels).

```python
def answer_three():
    cancerdf = answer_one()
    
    # Your code here
    #split data into X -> the data and y -> the labels(target) using iloc 
    X = cancerdf.iloc[:, 0:30]
    y = cancerdf.iloc[:, 30]
    
    return X, y

answer_three()
```

### Question 4
Using `train_test_split`, split `X` and `y` into training and test sets

```python
from sklearn.model_selection import train_test_split

def answer_four():
    X, y = answer_three()
    
    # Your code here
    # split X and y into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
    
    
    return X_train, X_test, y_train, y_test

answer_four()
```

### Question 5
Using KNeighborsClassifier, fit a k-nearest neighbors (knn) classifier with `X_train`, `y_train`.

```python
from sklearn.neighbors import KNeighborsClassifier

def answer_five():
    X_train, X_test, y_train, y_test = answer_four()
    
    # Your code here
    # instance of the classifier
    knn = KNeighborsClassifier(n_neighbors = 1)
    # train the classifier (fit the estimator) using the training data
    knn.fit(X_train, y_train)
    
    return  knn# Return your answer

answer_five()
```

### Question 6
Using your knn classifier, predict the class label using the median value for each feature.

```python

def answer_six():
    cancerdf = answer_one()
    
    # Your code here
    knn = answer_five()
    
    features_name = cancerdf.columns[0:30].tolist()
    # calculate median value of each feature
    cancerdf_median = cancerdf[features_name].median()
    cancerdf_median = cancerdf_median.values.reshape(1,-1)

    predict1 = knn.predict(cancerdf_median)
    
    
    return predict1# Return your answer

answer_six()
```

### Question 7
Using your knn classifier, predict the class labels for the test set `X_test`.

```python
def answer_seven():
    X_train, X_test, y_train, y_test = answer_four()
    knn = answer_five()
    
    # Your code here
    # predict the class labels for the test set X_test
    predict2 = knn.predict(X_test)
    
    return predict2 # Return your answer

answer_seven()
```

### Question 8
Find the score (accuracy) of your knn classifier using `X_test` and `y_test` & `X_train` and `y_train`. Do you notice overfitting or underfitting?

```python
from sklearn.metrics import accuracy_score

def answer_eight():
    X_train, X_test, y_train, y_test = answer_four()
    knn = answer_five()
    
    # Your code here
    train_score = knn.score(X_train, y_train)
    test_score = knn.score(X_test, y_test)
    
    if train_score > test_score:
        print('Overfitting')
    elif train_score < test_score:
        print('Underfitting')
    elif train_score == test_score:
        print('Idealfitting')
    
    return train_score, test_score # Return your answer

answer_eight()

#Overfitting as train score bigger that test score 
```

### Question 9
##### Accuracy Plot
* Create `mal_train_X`, `mal_train_y`,`ben_train_X`,`ben_train_y`, `mal_test_X`,`mal_test_y`, `ben_test_X` and `ben_test_y` portions of the dataset. Where `mal_train_X` contains all training samples that belongs to malignant class while `mal_train_y` contains all labels of the samples, and so on for other variables.

* Calculate Accuracy scores for each case and save all the results in a list called `scores` using knn model from question 5.

Try using the plotting function below to visualize the differet predicition scores between training and test sets, as well as malignant and benign cells.

```python
def accuracy_plot():
    import matplotlib.pyplot as plt

    %matplotlib notebook

    X_train, X_test, y_train, y_test = answer_four()

    # Find the training and testing accuracies by target value (i.e. malignant, benign)
    
    # #################################################################################
    # ###################### Write your code here #####################################
    # #################################################################################
    
    mal_train_X = X_train[y_train == 0]
    mal_train_y = y_train[y_train == 0]
    
    ben_train_X = X_train[y_train == 1]
    ben_train_y = y_train[y_train == 1]
    
    mal_test_X = X_test[y_test == 0]
    mal_test_y = y_test[y_test == 0]
    
    ben_test_X = X_test[y_test == 1]
    ben_test_y = y_test[y_test == 1]

    knn = answer_five()
    
    mal_train_score = knn.score(mal_train_X, mal_train_y)
    ben_train_score = knn.score(ben_train_X, ben_train_y)
    mal_test_score = knn.score(mal_test_X, mal_test_y)
    ben_test_score = knn.score(ben_test_X, ben_test_y)
    
    scores = [mal_train_score, ben_train_score, mal_test_score, ben_test_score]
    
    plt.figure()

    # Plot the scores as a bar chart
    bars = plt.bar(np.arange(4), scores, color=['#4c72b0','#4c72b0','#55a868','#55a868'])

    # directly label the score onto the bars
    for bar in bars:
        height = bar.get_height()
        plt.gca().text(bar.get_x() + bar.get_width()/2, height*.90, '{0:.{1}f}'.format(height, 2), 
                     ha='center', color='w', fontsize=11)

    # remove all the ticks (both axes), and tick labels on the Y axis
    plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

    # remove the frame of the chart
    for spine in plt.gca().spines.values():
        spine.set_visible(False)

    plt.xticks([0,1,2,3], ['Malignant\nTraining', 'Benign\nTraining', 'Malignant\nTest', 'Benign\nTest'], alpha=0.8);
    plt.title('Training and Test Accuracies for Malignant and Benign Cells', alpha=0.8)
```

```python
accuracy_plot() 
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

# load Iris dataset
iris = load_iris()
iris.keys()
```

```python
print(iris.DESCR)
```

```python
# create X DataFrame -> data and y DataFrame -> target
X = pd.DataFrame(iris.data, columns = iris.feature_names, dtype = float)
y = pd.DataFrame(iris.target, columns = ['target'], dtype = int)
```

```python
# Split the data into training and testing (30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# plotting a scatter matrix
# importing colour map
from matplotlib import cm

cmap = cm.get_cmap('gnuplot')
scatter = pd.plotting.scatter_matrix(X_train, c=y_train['target'] ,marker = 'o', s=40, hist_kwds={'bins':15}, figsize=(10,10), cmap=cmap)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# choose petal witdth and petal length features to train knn model
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train.iloc[:, 2:4], y_train['target'])
```

```python
# get score (accuracy) of knn classifier using X_test and y_test & X_train and y_train of the two selected features
train_score = knn.score(X_train.iloc[:, 2:4], y_train)
test_score = knn.score(X_test.iloc[:, 2:4], y_test)
print('Train Score (Accuracy) using Petals is:', train_score)
print('Test Score (Accuracy) using Petals is:', test_score)

if train_score > test_score:
    print('Overfitting')
elif train_score < test_score:
    print('Underfitting')
elif train_score == test_score:
    print('Idealfitting')
```

```python
def accuracy_plot():
    import matplotlib.pyplot as plt

    %matplotlib notebook

    # Find the training and testing accuracies by target value (i.e. Setosa, Versicolour, Virginica)

    # #################################################################################
    # ###################### Write your code here #####################################
    # #################################################################################
    X_Sepal_train = X_train.iloc[:, 2:4]
    X_Sepal_test = X_test.iloc[:, 2:4]

    set_train_X = X_Sepal_train[y_train['target'] == 0]
    set_train_y = y_train[y_train['target'] == 0]

    vers_train_X = X_Sepal_train[y_train['target'] == 1]
    vers_train_y = y_train[y_train['target'] == 1]

    virg_train_X = X_Sepal_train[y_train['target'] == 2]
    virg_train_y = y_train[y_train['target'] == 2]

    set_test_X = X_Sepal_test[y_test['target'] == 0]
    set_test_y = y_test[y_test['target'] == 0]

    vers_test_X = X_Sepal_test[y_test['target'] == 1]
    vers_test_y = y_test[y_test['target'] == 1]

    virg_test_X = X_Sepal_test[y_test['target'] == 2]
    virg_test_y = y_test[y_test['target'] == 2]

    set_train_score = knn.score(set_train_X, set_train_y)
    vers_train_score = knn.score(vers_train_X, vers_train_y)
    virg_train_score = knn.score(virg_train_X, virg_train_y)

    set_test_score = knn.score(set_test_X, set_test_y)
    vers_test_score = knn.score(vers_test_X, vers_test_y)
    virg_test_score = knn.score(virg_test_X, virg_test_y)


    scores = [set_train_score, vers_train_score, virg_train_score, set_test_score, vers_test_score, virg_test_score]

    plt.figure()

    # Plot the scores as a bar chart
    bars = plt.bar(np.arange(6), scores, color=['#4c72b0','#4c72b0','#55a868','#55a868', '#E50000', '#E50000'])

    # directly label the score onto the bars
    for bar in bars:
        height = bar.get_height()
        plt.gca().text(bar.get_x() + bar.get_width()/2, height*.90, '{0:.{1}f}'.format(height, 2), 
                     ha='center', color='w', fontsize=11)

    # remove all the ticks (both axes), and tick labels on the Y axis
    plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

    # remove the frame of the chart
    for spine in plt.gca().spines.values():
        spine.set_visible(False)

    plt.xticks([0,1,2,3,4,5], ['Setosa\nTraining', 'Versicolour\nTraining', 'Virginica\nTraining', 'Setosa\nTest', 'Versicolour\nTest', 'Virginica\nTest'], alpha=0.8);
    plt.title('Training and Test Accuracies for Setosa, Versicolour and Virginica', alpha=0.8)
```
